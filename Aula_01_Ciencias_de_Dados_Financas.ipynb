```python
data['Lucro_Liquido_12m'] = data['Lucro_Liquido_12m'].astype('int')
data['Patrim_Liq'] = data['Patrim_Liq'].astype('int')
data['PL'] = data['PL'].astype('float') / 100
```

```python
#  ROE =  Lucro Liquido do Ano / Patrimonio Liquido
def get_roe(data):
    return data["Lucro_Liquido_12m"]/data["Patrim_Liq"]
```
